```python
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
study = np.array([1,2,3,4,5,6,7])    # Study hour
gpa = np.array([1,1.3,2.5,2.6,3.5,3.7,4]) # gpa
sbn.regplot(x=study, y=gpa, ci=None, order=1)
plt.title("Study Hours Vs GPA of Students")
plt.xlabel("Study Hours")
plt.ylabel("GPA")
plt.show()
```

```python
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
drug = np.array([0,1,2,3,4,5,6,7])    # Study hour
forgetness = np.array([1.86,1.31,0.62,0.33,0.09,-0.67, -1.23,-1.37]) # gpa
sbn.regplot(x=drug, y=forgetness, ci=None, order=1)
plt.title("Clinical Trial")
plt.xlabel("Drug dosage (ml)")
plt.ylabel("Forgetfulness")
plt.show()
```
