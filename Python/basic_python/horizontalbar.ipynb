```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = [450,500,300,360,400,540]
plt.scatter(x,x, label='Data')
plt.bar(np.mean(x), np.mean(x),label='Mean')
# plt.axhline(y=np.mean(x),label='Mean', color='y')
plt.xlabel("Student's Marks")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```
