# To Dsplay Multiple plot's in a single figure
#### With the subplot() function you can draw multiple plots in one figure

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
import random
# subplot(nrows, ncols,index)
x = [1,2,3,4,5]
y = ['a','b','c','d','e']
plt.subplot(2,5,1)
plt.plot(x,y)
plt.show()
x1 = [1,2,3,4,5]
plt.subplot(2,5,2)
plt.pie(x1)
plt.show()
x2 = [2,3,4,5,6,7]
y2 = [2,3,4,5,6,7]
plt.subplot(2,5,3)
plt.subplot(2,5,2)
plt.plot(x2,y2)
plt.show()
x3 = [3,4,5,6,7,8]
y3 = [3,4,5,6,7,8]
plt.subplot(2,5,3)
plt.plot(x3,y3)
plt.show()
```
