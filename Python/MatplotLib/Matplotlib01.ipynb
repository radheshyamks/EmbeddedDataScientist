```python
import matplotlib.pyplot as plt;
temp = [8.5,10.5,6.8]
date = ['25/12','26/12','27/12']
plt.plot(date, temp)
plt.ylabel("Temperature")
plt.xlabel("Date")
plt.title("Date wise Temperature")
plt.show()
```

```python
import matplotlib.pyplot as plt;
temp = [8.5,10.5,6.8]
date = ['25/12','26/12','27/12']
plt.plot(date, temp)
plt.xlabel("Date")
plt.ylabel("Temperaturre")
plt.title("Date wise Temperature")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt;
temp = [8.5,10.5,6.8]
date = ['25/12','26/12','27/12']
plt.plot(date, temp, marker='o')
plt.xlabel("Date")
plt.ylabel("Temperaturre")
plt.title("Date wise Temperature")
plt.grid(True)
plt.show()
```
